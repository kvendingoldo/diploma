```python
from sympy import *
from IPython.display import display

# misc
k, l = symbols("k l")
M = Symbol("M")

# переменные
x1, x2, t = var('x1 x2 t', real=True)

# константы
rho, rho_a, gamma, theta = symbols('rho rho_a gamma theta')
f, W, g, c, Pa, h = symbols('f W g c Pa h')
e11, e12, e22 = symbols('epsilon_11, epsilon_12, epsilon_22')
H0 = Symbol('H0')

#rho = 1000
#c = 10
#g = 9.832
#gamma = 0.04472135955
#W = 10
theta = 0
#Pa = 100
#rho_a = 1.2754
f = 0
#f = 0.973 * 10e-4
h = 0

e11 = 0
e12 = 0
e22 = 0

# базисные функции
N_k = Function('N_k') 
W_l = Function('W_l')

# коэффициенты
a_k, a_mk, a_2mk = symbols('a_k, a_{m+k}, a_{2m+k}')

# функции
#q1 = Sum(a_k(t) * N_k(x1, x2, k), (k, 1, M))
#q2 = Sum(a_mk(t) * N_k(x1, x2, k), (k, 1, M))
#H = Sum(a_2mk(t) * N_k(x1, x2, k), (k, 1, M))
H, q1, q2 = symbols('H q1 q2')

# выражения
N_11 = 2 * e11 * Derivative(q1, x1)
N_12 = e12 * (Derivative(q1, x2) + Derivative(q2, x1))
N_22 = 2 * e22 * Derivative(q2, x2)
N_p = rho * g * H**2 / 2 + H * Pa
#N_11, N_22, N_12, N_p = symbols('N_11 N_22 N_12 N_p')


init_printing(use_latex='mathjax')

# уравнения

eq1 = Derivative(q1, t) - Derivative((N_11 - N_p), x1) - Derivative(N_12, x2) - f * q2 - (gamma**2 * rho_a * W**2 * cos(theta)) + (g/c**2) * (1/rho) * (q1 * sqrt(q1**2 + q2**2) / H**2) - Pa * Derivative(H, x1) - (rho * g * H) * Derivative(h, x1)
#+Derivative((q1**2 / H), x1)
#+Derivative(((q1 * q2) / H), x2)

eq2 = Derivative(q2, t) - Derivative((N_22 - N_p), x2) - Derivative(N_12, x1) + f * q1  - (gamma**2 * rho_a * W**2 * sin(theta)) + (g/c**2) * (1/rho) * (q2 * sqrt(q1**2 + q2**2) / H**2) - Pa * Derivative(H, x2) - (rho * g * H) * Derivative(h, x2)
#+ Derivative(((q1 * q2) / H), x1) 
#+ Derivative((q2**2 / H), x2) 

eq3 = Derivative(q1, x1) + Derivative(q2, x2) + Derivative(rho * H, t)

```
